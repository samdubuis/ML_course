```python
def standardize(x):
    """Standardize the original data set."""
    mean_x = np.mean(x)
    x = x - mean_x
    std_x = np.std(x)
    x = x / std_x
    return x, mean_x, std_x
```

```python
_, algo_svd = dump.load("dump/default/dump_SVD")
_, algo_knn = dump.load("dump/default/dump_KNN")
_, algo_slopeone = dump.load("dump/default/dump_SlopeOne")
```

```python
array_SVD = np.ones((df2.shape[0],1))
array_KNN = np.ones((df2.shape[0],1))
array_SlopeOne = np.ones((df2.shape[0],1))
```

```python
for i in df2.iterrows():
    if i[0]%50000==0:
        print(i[0])
    array_SVD[i[0]] = algo_svd.estimate(int(i[1][2])-1, int(i[1][3])-1)
    array_KNN[i[0]]= algo_knn.estimate(int(i[1][2])-1, int(i[1][3])-1)[0]
    array_SlopeOne[i[0]]=algo_slopeone.estimate(int(i[1][2])-1, int(i[1][3])-1)
    
```

```python
def pred_svd():
    for i in df2.iterrows():
        if i[0]%10000==0:
            print("SVD :"+str(i[0]))
        array_SVD[i[0]] = algo_svd.estimate(int(i[1][2])-1, int(i[1][3])-1)

def pred_knn():
    for i in df2.iterrows():
        if i[0]%10000==0:
            print("knn :"+str(i[0]))
        array_KNN[i[0]] = algo_knn.estimate(int(i[1][2])-1, int(i[1][3])-1)[0]

def pred_slopeone():
    for i in df2.iterrows():
        if i[0]%10000==0:
            print("Slope :"+str(i[0]))
        array_SlopeOne[i[0]] = algo_slopeone.estimate(int(i[1][2])-1, int(i[1][3])-1)

        
from threading import Thread, current_thread 
thread1 = Thread(target=pred_svd)
thread2 = Thread(target=pred_knn)
thread3 = Thread(target=pred_slopeone)
thread1.start()
thread2.start()
thread3.start()
```

```python
X=np.concatenate((array_SVD,array_KNN,array_SlopeOne),axis=1)
np.save("pred_default_param_svd_knn_slope",X)
```

```python
X
X[np.where(X>5)]=5
X[np.where(X<1)]=1
```
